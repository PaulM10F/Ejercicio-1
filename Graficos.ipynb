```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV con los nombres de las columnas
df = pd.read_csv('recursos.csv', header=None, names=['FechaHora', 'Memoria (KB)', 'CPU (%)'])

# Mostrar las primeras filas para verificar que se han ledo correctamente
print(df.head())

# Convertir la columna 'FechaHora' a tipo datetime
df['FechaHora'] = pd.to_datetime(df['FechaHora'])

# Configurar la figura de la grfica
plt.figure(figsize=(14, 7))

# Graficar el consumo de Memoria
plt.subplot(2, 1, 1)  # 2 filas, 1 columna, 1 grfica
plt.plot(df['FechaHora'], df['Memoria (KB)'], label='Consumo de Memoria (KB)', color='b')
plt.xlabel('Fecha y Hora')
plt.ylabel('Memoria (KB)')
plt.title('Consumo de Memoria en el Tiempo')
plt.legend()
plt.grid(True)

# Graficar el consumo de CPU
plt.subplot(2, 1, 2)  # 2 filas, 1 columna, 2 grfica
plt.plot(df['FechaHora'], df['CPU (%)'], label='Consumo de CPU (%)', color='r')
plt.xlabel('Fecha y Hora')
plt.ylabel('CPU (%)')
plt.title('Consumo de CPU en el Tiempo')
plt.legend()
plt.grid(True)

# Ajustar el espaciado entre las grficas
plt.tight_layout()

# Mostrar la grfica
plt.show()
```
